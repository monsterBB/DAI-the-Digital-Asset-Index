```python
import json
import pandas as pd
import numpy as np
import os
```

```python
path = "/home/fabian/Opensea/data/events/20220306/events_all"
# change the working directory to the path where the data are located
os.chdir(path)
```

```python
# this list holds all the json filename
asset_collections = []

# creates a ScandirIterator aliased as files
with os.scandir(path) as files:
  # loops through each file in the directory
    for file in files:
        if file.name.endswith('.json'):
          # adds only the image files to the asset_collections list
            asset_collections.append(file.name)

```

```python
def event_collect_fun(asset_collections):
    events_all = pd.DataFrame()
    for assets in asset_collections:
        with open (assets) as json_data:
           event_json = json.load(json_data)
        event_df = pd.DataFrame(event_json)
        if event_df.empty==True:
            continue
        select_df = event_df[['auction_type', 'collection_slug', 'contract_address', 
                          'created_date', 'event_type', 'total_price']]
        asset_id = []
        owner = []
        token_id = []
        payment_tocken = []
        token_usd_price = []
        block_hash = []
        block_number = []
        from_account = []
        to_account = []
        transaction_hash = []
        transaction_timestamp = []
        seller_address = []
        seller_account = []
        winner_address = []
        winner_account = []
        for i in range(len(event_df)):
            try:
                asset_id.append(str(int(event_df.asset[i]['id'])))
            except:
                asset_id.append(None)
            try:
                owner.append(event_df.asset[i]['owner']['user']['username'])
            except:
                owner.append(None)
            try:
                token_id.append(event_df.asset[i]['token_id'])
            except:
                token_id.append(None)
            try:
                payment_tocken.append(event_df.payment_token[i]['symbol'])
            except:
                payment_tocken.append(None)
            try:
                token_usd_price.append(event_df.payment_token[i]['usd_price'])
            except:
                token_usd_price.append(None)
            try:
                block_hash.append(event_df.transaction[i]['block_hash'])
            except:
                block_hash.append(None)
            try:
                block_number.append(event_df.transaction[i]['block_number'])
            except:
                block_number.append(None)
            try:
                from_account.append(event_df.transaction[i]['from_account']['address'])
            except:
                from_account.append(None)
            try:
                to_account.append(event_df.transaction[i]['to_account']['address'])
            except:
                to_account.append(None)
            try:
                transaction_hash.append(event_df.transaction[i]['transaction_hash'])
            except:
                transaction_hash.append(None)
            try:
                seller_address.append(event_df.seller[i]['address'])
            except:
                seller_address.append(None)
            try:
                seller_account.append(event_df.seller[i]['user']['username'])
            except:
                seller_account.append(None)
            try:
                winner_address.append(event_df.winner_account[i]['address'])
            except:
                winner_address.append(None)
            try:
                winner_account.append(event_df.winner_account[i]['user']['username'])
            except:
                winner_account.append(None)    
            try:
                transaction_timestamp.append(event_df.transaction[i]['timestamp'])
            except:
                transaction_timestamp.append(None)
        select_df['id'] = asset_id        
        select_df['owner'] = owner
        select_df['token_id'] = token_id 
        select_df['payment_tocken'] = payment_tocken
        select_df['token_usd_price'] = token_usd_price
        select_df['block_hash'] = block_hash
        select_df['block_number'] = block_number
        select_df['from_account'] = from_account
        select_df['to_account'] = to_account
        select_df['transaction_hash'] = transaction_hash
        select_df['transaction_time'] = transaction_timestamp
        select_df['seller_address'] = seller_address
        select_df['seller_account'] = seller_account
        select_df['winner_address'] = winner_address
        select_df['winner_account'] = winner_account
        events_all = pd.concat([events_all, select_df], ignore_index=True)
    return(events_all)
```

```python
events_all = event_collect_fun(asset_collections)
```

```python
path = "/home/fabian/Opensea/data/assets/20220306"
# change the working directory to the path where the data are located
os.chdir(path)
```

```python
# this list holds all the json filename
asset_collections = []

# creates a ScandirIterator aliased as files
with os.scandir(path) as files:
  # loops through each file in the directory
    for file in files:
        if file.name.endswith('.json'):
          # adds only the image files to the asset_collections list
            asset_collections.append(file.name)
```

```python
def asset_collect_fun(asset_collections):
    assets_all = pd.DataFrame()
    for assets in asset_collections:
        with open (assets) as json_data:
            asset_json = json.load(json_data)
        asset_df = pd.DataFrame(asset_json)
        if asset_df.empty==True:
            continue
        select_df = asset_df[['id', 'token_id', 'image_url', 'image_preview_url', 'image_thumbnail_url', 'image_original_url', 
                              'external_link', 'num_sales', 'background_color', 'name', 'description', 'traits']]
        asset_address = []
        asset_contract_type = []
        created_date = []
        asset_name = []
        owner = []
        schema_name = []
        owner_address = []
        creator_adress = []
        trait_type = []
        trait_value = []
        collection_slug = []
        collection_n = assets.split('_', 1)[0]
        for i in range(len(asset_df)):
            try:
                asset_address.append(asset_df.asset_contract[i]['address'])
            except:
                asset_address.append(None)
            try:
                asset_contract_type.append(asset_df.asset_contract[i]['asset_contract_type'])
            except:
                asset_contract_type.append(None)
            try:
                created_date.append(asset_df.asset_contract[i]['created_date'])
            except:
                created_date.append(None)
            try:
                asset_name.append(asset_df.asset_contract[i]['name'])
            except:
                asset_name.append(None)
            try:
                owner.append(asset_df.asset_contract[i]['owner'])
            except:
                owner.append(None)
            try:
                schema_name.append(asset_df.asset_contract[i]['schema_name'])
            except:
                schema_name.append(None)
            try:
                owner_address.append(asset_df.owner[i]['address'])
            except:
                owner_address.append(None)
            try:
                creator_adress.append(asset_df.creator[i]['address'])
            except:
                creator_adress.append(None)
            try:
                trait = asset_df.traits[i]
                t_type = ''
                t_value = ''
                for t_object in trait:
                    t_type = t_type + ',' + t_object['trait_type']
                    t_value = t_value + ',' + t_object['value']
                trait_type.append(t_type[1:])
                trait_value.append(t_value[1:])
            except:
                trait_type.append(None)
                trait_value.append(None)
            collection_slug.append(collection_n)
        select_df['asset_address'] = asset_address
        select_df['asset_contract_type'] = asset_contract_type
        select_df['created_date'] = created_date
        select_df['asset_name'] = asset_name
        select_df['owner'] = owner
        select_df['schema_name'] = schema_name
        select_df['owner_address'] = owner_address
        select_df['creator_adress'] = creator_adress
        select_df['trait_type'] = trait_type
        select_df['trait_value'] = trait_value
        select_df['collection_slug'] = collection_slug
        assets_all = pd.concat([assets_all, select_df], ignore_index=True)
    return(assets_all)
```

```python
assets_all = asset_collect_fun(asset_collections)
```
